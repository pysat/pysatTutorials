```python
import datetime as dt

import pysat
import pysatSpaceWeather
pysat.utils.registry.register_by_module(pysatSpaceWeather.instruments)
```

```python
# Date range for all examples
start_date = dt.datetime(2001, 1, 1)
end_date = dt.datetime(2001, 12, 31)
```

```python
# Create a Dst index Instrument
dst = pysat.Instrument('sw', 'dst')
```

```python
# Download data
dst.download(start_date, end_date)
```

```python
# Load a year of data
dst.load(date=start_date, end_date=end_date)
dst['dst'].plot(title=dst.meta['dst', dst.meta.labels.name], 
                ylabel=dst.meta['dst', dst.meta.labels.units])
```

```python
# Load a 'short' span of data
dst.load(2001, 1, 2001, 12)
dst['dst'].plot(title=dst.meta['dst', dst.meta.labels.name], 
                ylabel=dst.meta['dst', dst.meta.labels.units])
```

```python
# Load a single day
dst.load(2001, 1)
dst['dst'].plot(title=dst.meta['dst', dst.meta.labels.name], 
                ylabel=dst.meta['dst', dst.meta.labels.units])
```
